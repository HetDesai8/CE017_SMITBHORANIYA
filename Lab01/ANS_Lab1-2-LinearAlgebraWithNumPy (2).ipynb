```python
tweet = positive[2277]  #selected sample
print(tweet)
```

```python
nltk.download('stopwords')
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer 
#Removing Hyperlink
print('\033[92m'+tweet)
tweet2 = re.sub(r'https?:\/\/.*[\r\n]*','',tweet)
#removing hashtag
tweet2 = re.sub(r'#','',tweet2)
print('\033[94m', tweet2)
```

```python
print('\033[92m'+tweet2)

tokenizer = TweetTokenizer(preserve_case=False)
tweet_tokens = tokenizer.tokenize(tweet2)
print('\033[94m')
print('Tokenized String:')
print(tweet_tokens)
```

```python
#Remving stopwords and punctuation
stopwords_english = stopwords.words('english')
print('StopWords : \n ' , stopwords_english)
print('Punctuation : \n' , string.punctuation,'\n')

print('\033[92m',tweet_tokens)
print('\033[94m', 'Tweets Clean :')
tweet_clean = []

for word in tweet_tokens:
  if (word not in stopwords_english and word not in string.punctuation):
    tweet_clean.append(word)

print(tweet_clean)    
```

```python
print('\033[92m',tweet_clean)
print('\033[94m', 'Tweets Steam :')
tweet_stem =[]
stemmer = PorterStemmer()
for word in tweet_clean:
  stem_word = stemmer.stem(word)
  tweet_stem.append(stem_word)

print(tweet_stem)  
```
